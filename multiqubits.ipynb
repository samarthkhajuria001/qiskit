```python
import qiskit
# from qiskit import *
import numpy as np 
from qiskit.visualization import plot_histogram as histogram
from qiskit.visualization import plot_bloch_multivector as bloch
from qiskit_textbook.tools import array_to_latex
from math import sqrt,pi
```

```python
circ = qiskit.QuantumCircuit(2)
circ.draw('mpl')
```

```python
for i in range(0,2):
    circ.h(i)
circ.draw('mpl')
```

```python
simulator = qiskit.Aer.get_backend('statevector_simulator')
finalstate = qiskit.execute(circ,simulator).result().get_statevector()
print(finalstate)
array_to_latex(finalstate, pretext="\\text{Statevector} = ")
bloch(finalstate)
```

```python
initialstate = [1/sqrt(2),1j/sqrt(2)]
qc = qiskit.QuantumCircuit(2)
qc.initialize([1,0],0)
qc.initialize(initialstate,1)
qc.draw('mpl')

        
```

```python
statevector1 = qiskit.execute(qc,simulator).result().get_statevector()
```
