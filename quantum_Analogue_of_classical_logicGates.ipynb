```python
import qiskit 
from qiskit.visualization import plot_histogram,plot_bloch_multivector
from qiskit_textbook.tools import array_to_latex
```

# Logical AND Gate

```python
def logicand(x,y):
    qc_and = qiskit.QuantumCircuit(3)
    if x==1:
        x_in=[0,1]
    else:
        x_in=[1,0]
    if y==1:
         y_in=[0,1]
    else:
         y_in=[1,0]
    qc_and.initialize(x_in,0)
    qc_and.initialize(y_in,1)
    qc_and.ccx(0,1,2)
    return qc_and

    
```

```python
qc_and=logicand(1,0)
qc_and.draw('mpl')
```

```python
qc_and.measure_all()
```

```python
simulator = qiskit.Aer.get_backend('statevector_simulator')
result = qiskit.execute(qc_and,simulator).result()
plot_bloch_multivector(result.get_statevector())
```

# logical OR Gate

```python
def logicOr(x,y):
    qc = qiskit.QuantumCircuit(3)
    if x==0:
        x_in=[0,1]
    else:
        x_in=[1,0]
    if y==0:
         y_in=[0,1]
    else:
         y_in=[1,0]
    qc.initialize(x_in,0)
    qc.initialize(y_in,1)
    qc.initialize([0,1],2)
    qc.ccx(0,1,2)
    return qc
```

```python
qc_or = logicOr(0,1)
or_statevector = qiskit.execute(qc_or,simulator).result().get_statevector()
plot_bloch_multivector(or_statevector)
```
