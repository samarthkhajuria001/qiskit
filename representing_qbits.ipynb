```python
import qiskit
import numpy as np
from qiskit.tools.visualization import plot_histogram
```

```python

circ = qiskit.QuantumCircuit(3)
initial_state = [0,1]
circ.draw('mpl')
```

```python
circ.initialize(initial_state,1)
circ.initialize(initial_state,0)
circ.draw('mpl')
```

```python
backend = qiskit.Aer.get_backend('statevector_simulator')
```

```python
result = qiskit.execute(circ,backend).result()
```

```python
out_state = result.get_statevector()
```

```python
print(out_state)

#qiskit initialize q0 as <2|, q1 as <1| and q2 as <0|
```

```python
plot_histogram(result.get_counts())
```

```python
circ.measure_all()
circ.draw('mpl')
```

```python
result = qiskit.execute(circ,backend).result()
count = result.get_counts()
```

```python
print(count)
plot_histogram(count)
```

 Quick Exercise 

1. Create a state vector that will give a  1/3  probability of measuring  |0⟩ .
2. Create a different state vector that will give the same measurement probabilities.
3. Verify that the probability of measuring  |1⟩  for these two states is  2/3 .

```python
from math import sqrt
circ1 = qiskit.QuantumCircuit(1)
simulator = qiskit.Aer.get_backend('statevector_simulator')
initial_state = [1/sqrt(3), sqrt(2/3)]
circ1.initialize(initial_state,0)
result1 = qiskit.execute(circ1,simulator).result()
statevector = result1.get_statevector()
print(statevector)
print(result1.get_counts())
```

```python
plot_histogram(result1.get_counts())
```

```python
circ1.measure_all()
circ1.draw('mpl')
```

```python

result2 = qiskit.execute(circ1,simulator).result()
statevector2 = result2.get_statevector()
print(str(statevector2))
```

```python
plot_histogram(result2.get_counts())
```
