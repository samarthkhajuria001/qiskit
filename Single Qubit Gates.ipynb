```python
import qiskit
from math import pi,sqrt
from qiskit.visualization import plot_bloch_multivector
```

```python
circ = qiskit.QuantumCircuit(1)
circ.draw('mpl')
```

```python
# X-Gate

circ.x(0)
circ.draw('mpl')
```

```python
simulator = qiskit.Aer.get_backend('statevector_simulator')
result = qiskit.execute(circ,simulator).result()
statevectorX= result.get_statevector()
```

```python
print(statevectorX)
plot_bloch_multivector(statevectorX)
```
